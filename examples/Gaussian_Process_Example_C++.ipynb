# C++ Gaussian Process Examples

```python
import CMGDB

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import time
import math
import csv
```

# 2D ODE model

```python
# Load data from file
data = np.loadtxt('PO_50.dat')
# data = np.loadtxt('PO_100.dat')

# Get X and Y vectors
X = data[:, [0,1]]
Y = data[:, [2,3]]
```

```python
# Plot data
plt.plot(X[:,0], X[:,1], 'ro', Y[:,0], Y[:,1], 'b*');
```

# With a uniform grid

```python
phase_subdiv = 20
lower_bounds = [-2.0, -2.0]
upper_bounds = [2.0, 2.0]

model = CMGDB.Model(phase_subdiv, lower_bounds, upper_bounds, X, Y)
```

```python
%%time
morse_graph, map_graph = CMGDB.ComputeMorseGraph(model)
```

```python
CMGDB.PlotMorseGraph(morse_graph)
```

```python
CMGDB.PlotMorseSets(morse_graph, xlim=[-2, 0], ylim=[-1, 2])
```

# With an adaptive grid

```python
subdiv_min = 20
subdiv_max = 30
lower_bounds = [-2.0, -2.0]
upper_bounds = [2.0, 2.0]

model = CMGDB.Model(subdiv_min, subdiv_max, lower_bounds, upper_bounds, X, Y)
```

```python
%%time
morse_graph, map_graph = CMGDB.ComputeMorseGraph(model)
```

```python
CMGDB.PlotMorseGraph(morse_graph)
```

```python
CMGDB.PlotMorseSets(morse_graph, xlim=[-2, 0], ylim=[-1, 2])
```

# Leslie model

```python
# Load Leslie model data
# th1, th2 = 19.6, 23.68
data = np.loadtxt('Leslie_500.dat')
# data = np.loadtxt('Leslie_100.dat')

# Get X and Y vectors
X = data[:, [0,1]]
Y = data[:, [2,3]]
```

```python
# Plot data
plt.plot(X[:,0], X[:,1], 'ro', Y[:,0], Y[:,1], 'b*');
```

```python
subdiv_min = 20
subdiv_max = 30
lower_bounds = [-0.001, -0.001]
upper_bounds = [90.0, 70.0]

model = CMGDB.Model(subdiv_min, subdiv_max, lower_bounds, upper_bounds, X, Y)
```

```python
%%time
morse_graph, map_graph = CMGDB.ComputeMorseGraph(model)
```

```python
CMGDB.PlotMorseGraph(morse_graph)
```

```python
CMGDB.PlotMorseSets(morse_graph, xlim=[0, 60], ylim=[0, 40])
```
